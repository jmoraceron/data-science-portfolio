Solución 3.1-Calcular la media de la distribución de edades de la ciudad.

Paso1: Creamos la columna "edadxpoblacion" (Fi) en el data set multiplicando la "edad" por la "poblacion" (xi.fi)
Paso2: Hacemos la sumatoria Σ(xi.fi)
Paso3: Dividimos la suma de Σfi = N
Paso4: Calculamos la media usando la fórmula  X̅ = Σ(xi.fi)/N 

```r
sumxifi = sum(datanew$dataxifi)#Σ(xi.fi)
sumxifi
```

```r
sumfi = sum(datanew$datafi)#Σfi = N
sumfi
```

```r
media = sumxifi/sumfi#X̅ = Σ(xi.fi)/N 
media#media 44 años
```

Solución 3.2.-Calcular la desviación estándar de la distribución de edades de la ciudad. Para calcular la desviación estándar usamos la fórmula S=∑(xi−X̅)^2/n-1

Paso 1: Calculamos el cuadrado de la distancia a la media X̅ para cada dato xi(∑(xi−X̅)^2)
Paso 2: Sumamos los valores que resultaron del paso 2.
Paso 3: Dividimos entre el número de datos.

```r
for (i in 1:100){
    paso1 <- (i-media)^2
    print(paso1)}
```

```r
paso2 = sum(paso1)
paso2
```

```r
paso3 = paso2/100 # S=∑(xi−X̅)^2/n-1
paso3#desviación estándar 32 años
```
