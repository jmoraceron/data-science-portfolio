# Diseño del test de hipótesis

1. Definición de Hipótesis Nula y Alternativa.

Queremos validar que eL **sistema de recomendación** de producto a nuestros usuarios tenga un **efecto positivo**, es decir aumente el valor medio de las ventas de nuestra empresa.

**Hipótesis Nula (H0) mantiene que el valor medio de las ventas de nuestra empresa es menor adoptando el sistema de recomendación de producto a nuestros usuarios, mientras que la Hipótesis Alternativa(HA) plantea que el valor medio de las ventas de nuestra empresa ha aumentado adoptando el sistema de recomendación de producto a nuestros usuarios.

H0:  µ ventas en las que el sistema de recomendación ha participado <= µ ventas en las que el sistema de recomendación no ha participado.

HA:  µ ventas en las que el sistema de recomendación ha participado > µ ventas en las que el sistema de recomendación no ha participado.

2. Selección de estadístico de prueba.

Contamos en cada una de las columnas el número de testeos y encontramos:

**Número de ventas en las que el sistema de recomendación ha participado = 5502
**Número de ventas en las que el sistema de recomendación no ha participado = 6284

Como ambas tienen un número de testeos bastante mayor a 30 (la ley de los grandes números) y podemos encontrar el Valor de la media de H0, vamos a seleccionar como estadístico de prueba la **Distribución t** pero asumiendo una Distribución Gaussiana.


# 3. Validación y ejecución del test de hipótesis (2pts) 

- Como el tamaño de ambas muestras es suficientemente grande, resolvemos el contraste hallando el Z empírico para dos muestras independientes. 
- El nivel de significancia (Alfa) = 0.05
- El tamaño de la muestras es N > 30, entonces es suficientemente grande y conocemos las desviaciones estándar hacemos prueba t.
- Utilizaremos un test de una **cola positiva** porque lo que queremos validar es que el sistema de recomendación tenga un efecto positivo, por eso, la hipótesis nula (H0) afirma que el efecto de adoptar el sistema de recomendación de producto a nuestros usuarios es **menor**y la hipótesis alternativa (H1) afirma que el efecto es mayor.

```r
ventas_consist <- ventas_consistema$Order.Value
ventas_consist
```

```r
print("Z TEST DE UNA MUESTRA INDEPENDIENTE DE COLA POSITIVA PARA LA MEDIA ventas en las que el sistema de recomendación ha participado ")
tt = t.test(ventas_consist, 
       paired = F, 
       conf.level = 0.95, 
       alternative = "greater")
print(tt)
```

```r
n_grados = 11785
nivel_significancia = 0.05
# Valores críticos
valor_critico_superior = qt(1 - nivel_significancia, df = n_grados)
valor_critico_inferior = -valor_critico_superior
# Generamos y dibujamos una curva t
x = seq(-4, 4, length.out = 1000)
y = dt(x, df = n_grados)
plot(x, y, type = "l", lwd = 2, col = "blue", xlab = "Valor t", ylab = "Densidad", bty = "n")
# Pintamos el área bajo la curva 
x_area_superior = x[(x > valor_critico_superior) & (x <= max(x))]
y_area_superior = dt(x_area_superior, df = n_grados)
polygon(c(valor_critico_superior,x_area_superior,max(x_area_superior)),c(0, y_area_superior, 0),col = "lightgray", border = NA)
# Valores críticos
abline(v = c(valor_critico_superior), col = "#d94c9f", lty = 1)
# Añadimos valores críticos
text(valor_critico_superior + 0.2, 0.03, round(valor_critico_superior, 3), col = "black", font = 2)
# Título y leyenda
title("Curva t de Student con Zona de Rechazo", cex.main = 1.5)
legend("topright", legend = c("Curva t", "Zona de Rechazo"), fill = c("blue", "lightgray"), cex = 1, bty = "n")
```

 + En este escenario, el p valor es 0 que es menor que 0.05 entonces rechazamos la Hipótesis nula por lo tanto aceptamos la hipótesis Alternativa.
 + Intervalo de confianza del 95% para la diferencia en la media es [3115.827   TO  Inf]

```r
ventas_sinsist <- ventas_sinsistema$Order.Value
ventas_sinsist
```

```r
n_grados = 11785
nivel_significancia = 0.05
# Valores críticos
valor_critico_superior = qt(1 - nivel_significancia, df = n_grados)
valor_critico_inferior = -valor_critico_superior
# Generamos y dibujamos una curva t
x = seq(-4, 4, length.out = 1000)
y = dt(x, df = n_grados)
plot(x, y, type = "l", lwd = 2, col = "blue", xlab = "Valor t", ylab = "Densidad", bty = "n")
# Pintamos el área bajo la curva 
x_area_superior = x[(x > valor_critico_superior) & (x <= max(x))]
y_area_superior = dt(x_area_superior, df = n_grados)
polygon(c(valor_critico_superior,x_area_superior,max(x_area_superior)),c(0, y_area_superior, 0),col = "lightgray", border = NA)
# Valores críticos
abline(v = c(valor_critico_superior), col = "#d94c9f", lty = 1)
abline(v = c(tt$statistic), col = "#d94c9f", lty = 1)
# Añadimos valores críticos
text(valor_critico_superior + 0.2, 0.03, round(valor_critico_superior, 3), col = "black", font = 2)
# Título y leyenda
title("Curva t de Student con Zona de Rechazo", cex.main = 1.5)
legend("topright", legend = c("Curva t", "Zona de Rechazo"), fill = c("blue", "lightgray"), cex = 1, bty = "n")
```

 + En este escenario, el p valor es 0 que es menor que 0.05 entonces rechazamos la Hipótesis nula por lo tanto aceptamos la hipótesis Alternativa.

# Tamaño de la muestra para igualar las probabilidades de error tipo I y tipo II

Supongamos que necesitamos un error de tipo I y un error de tipo II igual a 0.05

Suponiendo que la desviación estándar de la muestra es igual a la desviación estándar de la población, podemos calcular el tamaño de muestra necesario de la siguiente manera:

1. La diferencia media de la hipótesis nula µ0 es 0.

2. La diferencia media de la hipótesis alternativa µ1 es 5.

3. La desviación estándar de la muestra es 2795.30319709415.

4. El valor α es 0.05

5. El valor β es 0.05, es decir, la potencia de la prueba es 0.95 = 95%

```r
power.t.test(power = .95, delta = 5, sd = sd(ventas_consist), 
             type = "t", 
             alternative = "one.sided", 
             sig.level = .05)
```
