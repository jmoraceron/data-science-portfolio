## Análisis descriptivo de la variable objetivo (target_deathrate)

El mínimo y el máximo de las medias de muertes por cancer es de 59.7 mínimo y 362.8 máximo cada 100K habitantes. El máximo de muertes por esta enfermedad en un condado representa el 0.362 % de su población, dato que podría ser para determinar si es un porcentaje bajo o alto de mortalidad.  

Tenemos una desviación estandar de 27.7 y una varianza de 770, como la media es de 178.6. Observamos una distribución normal en el gráfico histograma.

El valor que más se repite (moda) en la variable target es 184.3. 

En el boxplot observamos outliers tanto en el máximo como en el mínimo, el 25% de los valores son menores o igual a 161.2 y el 75% de los valores son menores o igual 195.2, la diferencia entre el valor del tercer cuartil y el primer cuartil es 34. 

La mediana se sitúa en el centro de la caja, la media y la mediana tienen valores muy cercanos y la moda no difiere mucho de estas, por lo tanto, la distribución es simétrica.

3.3 MODELO DE AJUSTE 
Ajustamos el modelo y obtenemos un resumen del modelo:

```r
modelo_lm = lm(tasas_mortalidad ~ cobertura_publica , data = data_train) # Entrenamos el modelo
summary(modelo_lm)
```

 - **coefficients:** nos dará los coeficientes estimados del modelo. En nuestro caso tendremos un coeficiente para la constante (el valor de Y cuando X es igual a cero)  y otro para la variable predictora (nos dice cuánto cambiará la variable dependiente por unidad de cambio en X). En nuestro caso nos indica que en promedio Y aumentará en **2.04572** unidades por cada unidad adicional de X. Por otro lado, observamos que cuando X sea 0 (solo con cobertura de salud proporcionada por el gobierno) el valor de las muertes por cancer sería aprox **139.30413** Además **Ambos coeficientes son estadísticamente poco significativos porque el valor p es menor a 0.05**
- **p.value:** en este caso estamos ante un modelo de regresión lineal simple, donde ya hemos visto que la variable "solo con cobertura de salud proporcionada por el gobierno" mantiene una relación poco significativa con "muertes por cancer".
- **df.residual:** nos indicará los grados de libertad residuales = núm.observaciones - núm.parámetros estimados >> 1827
- **sigma (residual standard error):** indica la desviación estándar de los residuos y por lo tanto el grado de dispersión de los residuos alrededor de la línea de regresión. En este caso, es 25.5
- **r.squared:** nos dice qué parte de la varianza de nuestra variable dependiente es explicada por nuestro modelo >> 0.1943
- **R-cuadrado ajustado**: 0.1938, esto muestra que solo el 19% de la variación disponible en el conjunto de datos dado (en "muertes por cancer") se explica por este modelo de regresión lineal simple. El 81% restante de variación de la variable "muertes por cancer" podría deberse a otros predictores o debido a una causa aleatoria.

Vamos a obtener diferentes atributos resultantes del modelo, por ejemplo:
- **residuals:** contiene los residuos del modelo >> las diferencias entre los valores observados y los valores predichos por el modelo.
- **fitted.values:** aquí encontraremos para cada registro los valores predichos por nuestro modelo
- **call**: nos recuerda la configuración de nuestro modelo

3.4 COMPROBAMOS EL RESTO DE SUPUESTOS: 
Normalidad: 
Visualización: Q-Q Plot de residuos.
Prueba estadística: la prueba de Shapiro-wilk sobre los residuos (muestras con pocas observaciones).
H0: La variable presenta una distribución normal
H1: La variable presenta una distribución no normal

Sig(p valor) > 0.05: No rechazar H0 (normal)
Sig(p valor) < 0.05: Rechazar H0

```r
# Visualization
qqnorm(modelo_lm$residuals,pch =19,col ="#8c0052", cex=2)
qqline(modelo_lm$residuals , col = "#3650b2", lwd = 2 )  

# Statistical Test
shapiro.test(modelo_lm$residuals)
```

El p-value es despreciable....
Dado que el valor p (0) < 0.05, por lo tanto, rechazamos la hipótesis nula, es decir, los errores no se distribuyen normalmente.

***Homocedasticidad:***
- Visualización: Residuos vs valores ajustados.

- Prueba estadística: la prueba de Breusch Pagan

```r
# Visualization
plot(modelo_lm$fitted.values , modelo_lm$residuals , pch =19,col ="#8c0052", cex=2 )

# Test Breusch-Pagan
bptest(modelo_lm)
```

El p-value es despreciable....
Dado que el valor p (0) < 0.05, por lo tanto, rechazamos la hipótesis nula, es decir, los errores no se distribuyen normalmente.

## Generad un gráfico de relación dato observado vs predicción

Elijo un modelo para la validación del modelo mediante el conjunto de Test ya que nos ayuda a determinar qué tan bien funcionará este modelo en datos.

Esto se hará mediante dos métodos:

- Visualización: el gráfico de tasas de mortalidad previstas frente a tasas de mortalidad reales para el conjunto de datos TEST.
- Medida estadística: R2, RMSE, MAE ayuda.

```r
# Making prediction
prediction = predict(modelo_lm, data_test)

### Visualization
plot(data_test$tasas_mortalidad, prediction, xlab="test.data$tasas_mortalidad", ylab="Prediction", pch =19,col ="#8c0052", cex=2)

#add fitted regression line
lm.out = lm(prediction ~ tasas_mortalidad, data = data_test)
newx = seq(min(data_test$tasas_mortalidad),max(data_test$tasas_mortalidad),by = 0.05)
conf_interval = predict(lm.out, newdata=data.frame(tasas_mortalidad=newx), interval="confidence", level = 0.95)
abline(lm.out, col = "#3650b2")

#add conf_interval lines
lines(newx, conf_interval[,2], col="#3650b2", lty=2)
lines(newx, conf_interval[,3], col="#3650b2", lty=2)

# Statistical Measure
data.frame( R2 = R2(prediction, data_test$tasas_mortalidad),
            RMSE = RMSE(prediction, data_test$tasas_mortalidad),
            MAE = MAE(prediction, data_test$tasas_mortalidad))
```
